# Easy
Выгрузить данные и отфильтровать их по какому-то значению.

Пример запроса с БД, приведенной выше: вывести учителей старше 30 лет.

```python
import pandas as pd
import sqlite3
```

```python
sql = sqlite3.connect('../03/mydb.db')
```

```python
sale_name = ["book", "dvd", "magazine", "game", "accesorize", "clothes", "food"]
salename_df = pd.DataFrame({'sale_name': sale_name})
sales_id = [1, 3, 0, 2, 5, 6, 4]
salename_df.to_sql('sale_name', sql, index_label="id", if_exists='replace') # Сохраним в базу
pd.read_sql("SELECT * FROM sale_name", sql)
```

```python
salesman = ["ivanov", "karasev", "gromova", "popov", "klimova", "tarasov", "petrov"]
age = [27, 22, 34, 35, 26, 31, 29]
salesman_df = pd.DataFrame({'salesman': salesman, 'age': age})
salesman_df.to_sql('salesman', sql, index_label="id", if_exists='replace')
pd.read_sql("SELECT * FROM salesman", sql)

salesman_id = [0, 1, 2, 3, 4, 5, 6]
salename_id = [6, 3, 2, 5, 1, 0, 4]
salesman_to_sale_df = pd.DataFrame({'salesman_id': salesman_id, 'salename_id': salename_id})
salesman_to_sale_df.to_sql("salesman_to_sale", sql, index_label="id", if_exists="replace")
pd.read_sql("SELECT * FROM salesman_to_sale", sql)
```

```python
price = [850, 1200, 370, 2500, 910, 1850, 560]
price_df = pd.DataFrame({'sale_name': sale_name, 'price': price})
price_df.to_sql('price', sql, index_label="id", if_exists="replace")
pd.read_sql("SELECT * FROM price", sql)

price_id = [0, 3, 2, 1, 5, 6, 4]
salename_id = [0, 1, 2, 3, 4, 5, 6]
price_to_sale_df = pd.DataFrame({'price_id': price_id, 'salename_id': salename_id})
price_to_sale_df.to_sql("price_to_sale", sql, index_label="id", if_exists="replace")
pd.read_sql("SELECT * FROM price_to_sale", sql)

price_df
```

```python
salesman_to_sale_df.set_index("salesman_id").join(salesman_df).set_index("salename_id").join(salename_df, rsuffix='salename', lsuffix='_salesman').reset_index(drop=True)
a_df = salesman_to_sale_df.join(salesman_df, on='salesman_id').join(salename_df, on='salename_id', rsuffix='salename', lsuffix='_salesman')
a_df
```

```python
a_df[a_df['age']>30]
```
